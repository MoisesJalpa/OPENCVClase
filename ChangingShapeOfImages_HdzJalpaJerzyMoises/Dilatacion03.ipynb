```python
#Dilatacion
#Importar librerias
import cv2
import matplotlib.pyplot as plt 

image = cv2.imread("rayos2.jpg")
#Tamaño de dilatacion
d1 = 0
d2 = 10
d3 = 20
#Tipo de dilatacion
t1 = cv2.MORPH_RECT
t2 = cv2.MORPH_CROSS     #Marcos
t3 = cv2.MORPH_ELLIPSE

tmp1 = cv2.getStructuringElement(t1, (2*d1 + 1, 2*d1+1), (d1, d1))
tmp2 = cv2.getStructuringElement(t2, (2*d2 + 1, 2*d2+1), (d2, d2))  #Segun el kernel se expande el pixel activo
tmp3 = cv2.getStructuringElement(t3, (2*d3 + 1, 2*d3+1), (d3, d3))
#Apply dilation to the images
final1 = cv2.dilate(image, tmp1)
final2 = cv2.dilate(image, tmp2)   #Aplicando dilatacion
final3 = cv2.dilate(image, tmp3)


plt.figure(0)
plt.imshow(final1)
plt.title("Dilatacion con estructura rectangular")
plt.axis('off')  

plt.figure(1)
plt.imshow(final2)
plt.title("Dilatacion con estructura  cruz")
plt.axis('off')

plt.figure(2)
plt.imshow(final3)
plt.title("Dilatacion con estructura eliptico")
plt.axis('off')

# Mostrar imagenes
plt.show()
```
