```python
#EROCION 
#Importando librerias
import cv2          #Lectura de imagenes
import matplotlib.pyplot as plt    #Visualizar graficos 

image = cv2.imread("rayos2.jpg")
#Define erosion size
s1 = 0      #un punto
s2 = 10
s3 = 10

t1 = cv2.MORPH_RECT                #Tipos de kernel
t2 = cv2.MORPH_CROSS
t3 = cv2.MORPH_ELLIPSE
#Define and save the erosion template
tmp1 = cv2.getStructuringElement(t1, (2*s1 + 1, 2*s1+1), (s1, s1))
tmp2= cv2.getStructuringElement(t2, (2*s2 + 1, 2*s2+1), (s2, s2))     #Marcos para estructuras
tmp3 = cv2.getStructuringElement(t3, (2*s3 + 1, 2*s3+1), (s3, s3))      # +1 para asegurar que haya un pixel central
#Apply the erosion template to the image and save in different variables
final1 = cv2.erode(image, tmp1)
final2 = cv2.erode(image, tmp2)         #Barrido sobre la imagen
final3 = cv2.erode(image, tmp3)
#Show all the images with different erosions
plt.figure(0)
plt.imshow(final1)
plt.title("Erosion con Estructura rectangular")
plt.axis('off')  # 
plt.figure(1)
plt.imshow(final2)
plt.title("Erosion con estructura en cruz")
plt.axis('off')

plt.figure(2)
plt.imshow(final3)
plt.title("Erosion Erosion con estructura eliptica")
plt.axis('off')

# Display all figures
plt.show()
```
